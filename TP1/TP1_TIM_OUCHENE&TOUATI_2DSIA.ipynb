## Creation de l'image couleur :

Dans cette partie, on n'a pas utilise l'instruction "cv.merge", mais on a pu creer chaque canal RGB seul et puis on a fait une synthese additive pour engender le reste des couleurs et puis on a fini par une concatenation.

```python
red = np.array([[255, 0, 0],[255, 0, 0],[255, 0, 0],[255, 0, 0]])
green = np.array([[0, 255, 0],[0, 255, 0],[0, 255, 0],[0, 255, 0]])
blue = np.array([[0, 0, 255],[0, 0, 255],[0, 0, 255],[0, 0, 255]])
```

# synthese additive et affichage :

```python
white = red + green + blue
```

```python
white
```

```python
yellow = red + green
yellow
```

```python
cyan = green + blue
cyan
```

```python
magenta = red + blue
magenta
```

```python
colors= np.concatenate((red, green, blue, white, yellow, cyan, magenta), axis=0)
colors
```

## Matrice qui represente l'image :

```python
image_1 = []
for i in range(28) :
    image_1.append(colors)
```

```python
image = np.asarray(image_1)
```

```python
image
```

## Conversion et affichage :

```python
image_CV = np.uint8(image)
#affichage:
plt.imshow(image_CV, cmap = 'gray')
plt.show()
```

# Bits de codage :
on a une variation de niveau de gris entre 0 et 255, alors on a de 0 a 2^8 - 1 cad : on a besoins de 8 bits.

# Poids theorique :
On a une matrice de dimension 28 * 28 * 3 pixels, et on a 7 couleurs couleurs, cad chaque couleur est sur 4 colonnes et 28 lignes, et puis chaque pixel et code sur 3 octes = 24 bits, alors on aura un nombre total de bits qui est de 28*28*3 octes, comme suit :

```python
image.shape
```

```python
poids_th = image.shape[0] * image.shape[1] * image.shape[2] * 8
print("le poids theorique est : ",poids_th, "bits    = ", poids_th/8, "octets" )
```

# Manip 3 :

## lecture et visualisation de l'image :

### Sans inversion des canaux :

```python
mandrill_img = cv.imread('mandrill.png')
plt.imshow(mandrill_img)
plt.show()
```

### Avec inversion des canaux :

```python
mandrill_img_rev = cv.cvtColor(mandrill_img,cv.COLOR_BGR2RGB)
plt.imshow(mandrill_img_rev)
plt.show()
```

### La taille et la profondeur de l’image :

```python
img_shape = mandrill_img_rev.shape 
img_dtype = mandrill_img_rev.dtype
print("la taille de l'image est : ", img_shape[0] * img_shape[1] * img_shape[2], " octets et sa profondeur est : ", img_dtype)
```

### Valeurs de type uint8 :

l'image est de type uint8 cad on code chaque pixel sur 8 bits, alors on peut avoir des variations de valeurs de chaque pixel de 0 a 255 : 256 = 2^^8 valeurs possibles

### Separation des canaux :

On utilise la commande "split" :

```python
canaux = cv.split(mandrill_img_rev)
```

```python
red_canal = canaux[0][:][:]
green_canal = canaux[1][:][:]
blue_canal = canaux[2][:][:]
```

### Affichage :

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(1,3,1)
plt.imshow(red_canal, cmap = "gray")
plt.title('Red canal', color = 'r')


fig.add_subplot(1,3,2)
plt.imshow(green_canal, cmap = "gray")
plt.title('Green canal', color = 'g')

fig.add_subplot(1,3,3)
plt.imshow(blue_canal, cmap = "gray")
plt.title('Blue canal', color = 'b')

plt.show()
```

# Explication :
On remarque bien les yeux du mandrill dans l'image originale apparaissent en couleur "orange", et on connait bien deja que l'orange est une synthese additive des couleurs "rouge" et "vert" avec des saturations differentes (on explique la saturation au niveau des canaux par la forte presence en blanc et absence totale en noir) - Ceci est bien montre dans l'image jointe -, ceci est aussi confirmee dans les differentes images des differents canaux car :
- Dans l'image du canal rouge, les yeux apparaissent avec une forte intensite(saturation) ce qui presente la forte presence du rouge (proche du 255)
- Et puis une faible intensite du vert (proche de 150)
- Et puis une tres tres faible intensite (on peut dire qu'elle est nulle) representee en couleur noir des yeux dans le canal bleu.
Toutes ces remarques sont confirmees par les valeurs de l'image suivante, ou on voit que l'orange est une synthese du rouge (presence totale = 255) et du vert (140 a 160) et absence du bleu (0) :


### interversion des canaux et affichages :

```python
RGB = np.dstack((red_canal, green_canal, blue_canal))
RBG = np.dstack((red_canal, blue_canal, green_canal))
GRB = np.dstack((green_canal, red_canal, blue_canal))
GBR = np.dstack((green_canal, blue_canal, red_canal))
BRG = np.dstack((blue_canal, red_canal, green_canal))
BGR = np.dstack((blue_canal, green_canal, red_canal))
```

```python
fig=plt.figure(figsize=(15, 20))

fig.add_subplot(1,3,1)
plt.imshow(RGB)
plt.title('RGB')

fig.add_subplot(1,3,2)
plt.imshow(RBG)
plt.title('RBG')

fig.add_subplot(1,3,3)
plt.imshow(GRB)
plt.title('GRB')

fig.add_subplot(2,3,1)
plt.imshow(GBR)
plt.title('GBR')

fig.add_subplot(2,3,2)
plt.imshow(BRG)
plt.title('BRG')

fig.add_subplot(2,3,3)
plt.imshow(BGR)
plt.title('BGR')
plt.show()
```

### Commentaires :

Un changement dans l'ordre des canaux implique un changement dans les couleurs visualisees, c'est vrai qu'il y a seulement 3 couleurs R, G et B mais chaque combinaison donne un resultat different des autres.

En utilisant la fonction predefinie :

```python
mandrill_NG = cv.cvtColor(mandrill_img_rev, cv.COLOR_BGR2GRAY)
plt.imshow(mandrill_NG, cmap = "gray")
plt.show()
```

### Comparaison :

```python
fig=plt.figure(figsize=(10, 10))

fig.add_subplot(1,2,1)
plt.imshow(luminance, cmap = "gray")
plt.title("A l'aide de l'expression donnee : ")

fig.add_subplot(1,2,2)
plt.imshow(mandrill_NG, cmap = "gray")
plt.title("A l'aide de la fonction predefinie : ")
plt.show()
```

On remarque qu'il y a une legere difference dans la clarte (luminance), ca se voit au niveau des yeux et nez :
- l'expression donne une image en NG plus claire.
- la fonction predefinie donne une image en NG plus foncee.
Ceci revient au coefficient de l'expression qui ne vont jamais donner le meme resultats (ce sont des valeurs proches), de plus il differe selon l'IDE qu'on travaille avec : 
- Cas de python : luminance= 0,2126 × Rouge + 0,7152 × Vert + 0,0722 × Bleu.
- Cas de MATLAB : Luminance = 0.2989 x R + 0.5870 x G + 0.1140 x B

Dans ce qui suit, on va confirmer nos resultats :

```python
img_diff = luminance - mandrill_NG
image_diff = np.uint8(img_diff)
#affichage:
plt.imshow(image_diff, cmap = 'gray')
plt.show()
```

Ce qui confirme bien que la fonction predefinie donne une image en NG differente que l'image donnee par l'expression de luminance.
